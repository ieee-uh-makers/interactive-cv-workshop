```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = [i for i in range(1, 11)]
yt = [0.2, 0.4, 0.55, 0.7, 0.8, 0.9, 0.95, 0.99, 0.999, 1]
tv = [0.15, 0.25, 0.3, 0.35, 0.35, 0.33, 0.3, 0.27, 0.25, 0.22]
plt.figure(dpi=150)
plt.plot(x, yt)
plt.plot(x, tv)
plt.grid()
plt.ylabel('Accuracy')
plt.xlabel('Training Epoch')
plt.annotate("Overfitting", [4, 0.35])
plt.legend(['Training', 'Validation'])
plt.savefig('overfit.jpg')
plt.show()
```
